````python
# float
salary = 75000.50
pi = 3.14159
print(type(salary))     # <class 'float'>
print(type(pi))
````

## 1.2 🔥 The Proof: Even numbers have methods!

````python
# This is the KEY insight - everything in Python is an object
# Even a simple number 10 has methods!

x = 10
print(type(x))           # <class 'int'> — it's an OBJECT of class int
print(dir(x))            # Look at ALL the methods a simple integer has!

# Some useful ones:
print((255).bit_length())   # How many bits needed to store 255? → 8
print((10).real)            # 10
print((10).imag)            # 0
````

````python
# Strings have tons of methods — because str is a CLASS
name = "hello world"
print(dir(name))   # Look at all these methods!

# You've been calling METHODS on OBJECTS this whole time
print(name.upper())       # method call on str object
print(name.title())       # method call on str object
print(name.split())       # method call on str object
print(name.replace('world', 'python'))  # method call on str object
````

## 1.3 Type Conversion (Casting)

````python
# int ↔ float ↔ str ↔ bool

# str to int
num_str = "42"
num = int(num_str)
print(num, type(num))        # 42 <class 'int'>

# int to float
x = float(10)
print(x, type(x))            # 10.0 <class 'float'>

# float to int (truncates!)
y = int(3.99)
print(y)                      # 3 (NOT 4 — it truncates, doesn't round)

# anything to bool
print(bool(0))        # False
print(bool(""))       # False
print(bool([]))       # False
print(bool(None))     # False
print(bool(1))        # True
print(bool("hello"))  # True
print(bool([1,2,3]))  # True

# RULE: 0, empty, None → False. Everything else → True
````

# PART 2 — Operators
*The tools you use to manipulate data*

## 2.1 Arithmetic Operators

````python
a, b = 17, 5

print(f"a + b  = {a + b}")     # 22   Addition
print(f"a - b  = {a - b}")     # 12   Subtraction
print(f"a * b  = {a * b}")     # 85   Multiplication
print(f"a / b  = {a / b}")     # 3.4  Division (always returns float)
print(f"a // b = {a // b}")    # 3    Floor division
print(f"a % b  = {a % b}")     # 2    Modulus (remainder)
print(f"a ** b = {a ** b}")    # 1419857  Power
````

## 2.2 Comparison Operators

````python
x, y = 10, 20

print(x == y)    # False  — Equal to
print(x != y)    # True   — Not equal to
print(x > y)     # False
print(x < y)     # True
print(x >= 10)   # True
print(x <= 5)    # False
````

## 2.3 Logical Operators

````python
age = 25
income = 50000

# and — both must be True
print(age > 18 and income > 30000)   # True

# or — at least one must be True
print(age > 30 or income > 30000)    # True

# not — flips the result
print(not age > 30)                   # True
````

## 2.4 Membership & Identity Operators

````python
# Membership: in, not in
fruits = ["apple", "banana", "mango"]
print("apple" in fruits)       # True
print("grape" not in fruits)   # True

name = "Python Programming"
print("Python" in name)        # True
````

````python
# Identity: is, is not (checks if SAME OBJECT in memory)
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a == b)     # True  — same VALUE
print(a is b)     # False — different OBJECTS in memory
print(a is c)     # True  — c points to the SAME object as a

# This matters in ML when comparing DataFrames and arrays!
````

## 4.3 break, continue, pass

````python
# break — exit the loop
for i in range(10):
    if i == 5:
        break
    print(i, end=" ")     # 0 1 2 3 4
print()

# continue — skip this iteration
for i in range(10):
    if i % 2 == 0:
        continue
    print(i, end=" ")     # 1 3 5 7 9
print()

# pass — do nothing (placeholder)
for i in range(5):
    pass  # TODO: implement later
````

## 4.4 List Comprehensions (🔥 Used EVERYWHERE in ML/AI)

````python
# Regular way
squares = []
for i in range(10):
    squares.append(i ** 2)
print(squares)

# List comprehension — same thing, one line
squares = [i ** 2 for i in range(10)]
print(squares)

# With condition
even_squares = [i ** 2 for i in range(10) if i % 2 == 0]
print(even_squares)   # [0, 4, 16, 36, 64]
````

````python
# Real ML example: cleaning data
raw_data = [" Alice ", "BOB", "  charlie", "DAVE  "]
clean_data = [name.strip().title() for name in raw_data]
print(clean_data)   # ['Alice', 'Bob', 'Charlie', 'Dave']
````

# PART 5 — Functions
*Reusable blocks of code*

## 5.1 Basic Functions

````python
def greet(name):
    return f"Hello, {name}!"

print(greet("Ravi"))     # Hello, Ravi!
print(greet("Priya"))    # Hello, Priya!
````

````python
# Default arguments
def power(base, exp=2):
    return base ** exp

print(power(5))       # 25 (default exp=2)
print(power(5, 3))    # 125
````

## 5.2 *args and **kwargs

````python
# *args — accepts any number of positional arguments
def add_all(*args):
    print(f"Received: {args}")   # it's a TUPLE
    return sum(args)

print(add_all(1, 2, 3))          # 6
print(add_all(10, 20, 30, 40))   # 100
````

````python
# **kwargs — accepts any number of keyword arguments
def print_info(**kwargs):
    print(f"Received: {kwargs}")   # it's a DICT
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

print_info(name="Ravi", age=25, city="Hyderabad")
````

## 5.3 Lambda Functions (🔥 Used in pandas, ML pipelines)

````python
# Regular function
def square(x):
    return x ** 2

# Lambda — same thing, one line
square = lambda x: x ** 2
print(square(5))     # 25

# Lambda with multiple arguments
add = lambda a, b: a + b
print(add(3, 7))     # 10
````

````python
# Real-world: sorting with lambda
students = [("Ravi", 85), ("Priya", 92), ("Amit", 78)]
students.sort(key=lambda x: x[1], reverse=True)
print(students)   # Sorted by marks descending

# Real-world: map + lambda
numbers = [1, 2, 3, 4, 5]
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)    # [2, 4, 6, 8, 10]

# Real-world: filter + lambda
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)      # [2, 4]
````

## 5.4 Scope: Local vs Global

````python
x = 100           # global

def my_func():
    x = 50         # local — different x!
    print(f"Inside function: x = {x}")

my_func()
print(f"Outside function: x = {x}")    # still 100
````

## 6.2 Tuples — Ordered, Immutable

````python
# Creating tuples
point = (10, 20)
rgb = (255, 128, 0)
single = (42,)       # Need comma for single element!

print(point[0])      # 10
print(len(rgb))      # 3

# Tuple unpacking — used ALL the time in ML
x, y = point
print(f"x={x}, y={y}")

# Multiple return values from function
def get_stats(numbers):
    return min(numbers), max(numbers), sum(numbers) / len(numbers)

mn, mx, avg = get_stats([10, 20, 30, 40, 50])
print(f"Min={mn}, Max={mx}, Avg={avg}")
````

## 6.3 Sets — Unordered, No Duplicates

````python
# Creating sets
nums = {1, 2, 3, 4, 5}
dupes = {1, 1, 2, 2, 3, 3}
print(dupes)              # {1, 2, 3} — duplicates removed!

# Set operations
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)      # Union:        {1, 2, 3, 4, 5, 6}
print(a & b)      # Intersection: {3, 4}
print(a - b)      # Difference:   {1, 2}
print(a ^ b)      # Symmetric:    {1, 2, 5, 6}

# Methods
nums.add(6)           # Add element
nums.remove(1)        # Remove (error if missing)
nums.discard(99)      # Remove (no error if missing)
print(nums)
````

````python
# ⚠️ SET LIMITATION: No indexing, no order, no key-value
s = {10, 20, 30}
try:
    print(s[0])       # This will CRASH — sets have no index
except TypeError as e:
    print(f"Error: {e}")
````

## 7.2 Your First Class — The Custom Data Type

````python
# WHY do we need this?
# Because list, tuple, dict, set — none of them can represent
# a STUDENT with data AND behavior bundled together.

class Student:
    
    def __init__(self, name, roll_no, marks):
        self.name = name           # attribute (data)
        self.roll_no = roll_no     # attribute (data)
        self.marks = marks         # attribute (data)
    
    def average(self):             # method (behavior)
        return sum(self.marks) / len(self.marks)
    
    def grade(self):               # method (behavior)
        avg = self.average()
        if avg >= 90: return 'A+'
        elif avg >= 80: return 'A'
        elif avg >= 70: return 'B'
        elif avg >= 60: return 'C'
        else: return 'F'


# Create OBJECTS (instances of the class)
s1 = Student("Ravi", 101, [85, 90, 78])
s2 = Student("Priya", 102, [92, 95, 88])

# Now data AND behavior live TOGETHER
print(f"{s1.name}: Average = {s1.average():.1f}, Grade = {s1.grade()}")
print(f"{s2.name}: Average = {s2.average():.1f}, Grade = {s2.grade()}")

# Check the type!
print(type(s1))    # <class '__main__.Student'> — YOUR custom class!
````

````python
# Compare: list/dict way vs class way

# ❌ OLD WAY — data and logic are separate
student_dict = {"name": "Ravi", "marks": [85, 90, 78]}
avg = sum(student_dict["marks"]) / len(student_dict["marks"])  # logic is outside

# ✅ NEW WAY — data and logic are bundled
s1 = Student("Ravi", 101, [85, 90, 78])
avg = s1.average()    # much cleaner! The object knows how to compute its own average
````

## 7.3 More Practical Example: BankAccount

````python
class BankAccount:
    
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
        self.transactions = []
    
    def deposit(self, amount):
        self.balance += amount
        self.transactions.append(f"Deposited ₹{amount}")
        print(f"₹{amount} deposited. Balance: ₹{self.balance}")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds!")
            return
        self.balance -= amount
        self.transactions.append(f"Withdrew ₹{amount}")
        print(f"₹{amount} withdrawn. Balance: ₹{self.balance}")
    
    def statement(self):
        print(f"\n--- Statement for {self.owner} ---")
        for t in self.transactions:
            print(f"  {t}")
        print(f"  Current Balance: ₹{self.balance}")


# Use it!
acc = BankAccount("Ravi", 1000)
acc.deposit(5000)
acc.withdraw(2000)
acc.withdraw(10000)    # Insufficient funds!
acc.deposit(3000)
acc.statement()
````

## 7.4 OOP Concepts

### Encapsulation — Protecting data

````python
class Employee:
    
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary     # __ makes it PRIVATE
    
    def get_salary(self):          # controlled access
        return self.__salary
    
    def give_raise(self, amount):
        if amount > 0:
            self.__salary += amount
            print(f"{self.name} got a raise! New salary: ₹{self.__salary}")


emp = Employee("Ravi", 50000)
print(emp.name)              # Works fine
print(emp.get_salary())      # 50000 — through method

# Can't access directly:
try:
    print(emp.__salary)      # This will CRASH — it's private!
except AttributeError as e:
    print(f"Error: {e}")

emp.give_raise(10000)
````

### Inheritance — Reusing code

````python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "Some sound"

class Dog(Animal):           # Dog INHERITS from Animal
    def speak(self):
        return "Woof!"

class Cat(Animal):           # Cat INHERITS from Animal
    def speak(self):
        return "Meow!"


animals = [Dog("Buddy"), Cat("Whiskers"), Dog("Max")]

for animal in animals:
    print(f"{animal.name} says {animal.speak()}")

# Buddy says Woof!
# Whiskers says Meow!
# Max says Woof!

print(isinstance(animals[0], Dog))      # True
print(isinstance(animals[0], Animal))   # True — Dog IS an Animal
````

### Polymorphism — Same method, different behavior

````python
# Polymorphism = "many forms"
# The SAME method name behaves differently for different classes

class Circle:
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14159 * self.radius ** 2

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height


shapes = [Circle(5), Rectangle(4, 6), Circle(3)]
for shape in shapes:
    print(f"{type(shape).__name__}: area = {shape.area():.2f}")

# Same method name .area() — different behavior!
````

### `__str__` and `__repr__` — Making objects printable

````python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def __str__(self):
        return f"{self.name} — ₹{self.price}"
    
    def __repr__(self):
        return f"Product('{self.name}', {self.price})"


p = Product("Laptop", 65000)
print(p)           # Calls __str__  → Laptop — ₹65000
print(repr(p))     # Calls __repr__ → Product('Laptop', 65000)
````

## 7.5 🎯 The Grand Reveal — Full Circle

````python
# ═══════════════════════════════════════════════════════
#  THE GRAND REVEAL
# ═══════════════════════════════════════════════════════

# Remember when you did this?
my_list = [1, 2, 3]
my_list.append(4)          # You called a METHOD on an OBJECT of class list

my_string = "hello"
my_string.upper()          # You called a METHOD on an OBJECT of class str

my_dict = {"a": 1}
my_dict.keys()             # You called a METHOD on an OBJECT of class dict

# Someone at Python HQ wrote:
#   class list:
#       def append(self, item): ...
#       def sort(self): ...
#
#   class str:
#       def upper(self): ...
#       def split(self): ...

# And NOW you can do the same!
class MyList:
    def __init__(self):
        self.data = []
    def add(self, item):
        self.data.append(item)
    def show(self):
        print(self.data)

ml = MyList()
ml.add(10)
ml.add(20)
ml.show()     # [10, 20]

print(type(ml))            # <class '__main__.MyList'>
print(type(my_list))       # <class 'list'>

# See? Your class is JUST LIKE Python's built-in classes.
# You're a class creator now. 🎉
````

**Q1.** Write a function `is_palindrome(text)` that returns True if a string reads the same forward and backward (ignore case).

````python
# YOUR CODE HERE
def is_palindrome(word):
    '''
        word = word.lower()
        rev = word[::-1]
        return word == rev
    '''
    word = word.lower()
    i, j = 0, len(word)-1
    while i<j:
        if word[i] == word[j]:
            pass
        else:
            return False
        i+=1
        j-=1
    return True
# Test:
# is_palindrome("Racecar") → True
# is_palindrome("Python") → False
print(is_palindrome("Racecar"))
print(is_palindrome("Python"))
````

**Q2.** Write a function `factorial(n)` using recursion.

````python
# YOUR CODE HERE
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)
print(factorial(5))
````

**Q3.** Write a function `count_vowels(text)` that returns the count of vowels in a string. Use a default parameter for what counts as vowels.

````python
# YOUR CODE HERE
def count_vowels(text, vowels = "aeiou"):
    text = text.lower()
    c = 0
    for i in text:
        if i in vowels:
            c+=1
    return c
print(count_vowels("Rgukt Basar"))
````

## Assignment 7: OOP — Classes & Objects (6 questions)

**Q1.** Create a `Rectangle` class with `width` and `height` attributes, and methods `area()`, `perimeter()`, and `is_square()`.

````python
# YOUR CODE HERE
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(self):
        return 2*(self.width + self.height)
    def is_square(self):
        return self.width == self.height
        
# Test:
# r = Rectangle(5, 5)
# r.area() → 25
# r.is_square() → True

r = Rectangle(5, 5)
print(r.area())
print(r.is_square())
````

**Q2.** Create a `BankAccount` class with `deposit()`, `withdraw()`, and `get_balance()` methods. The balance should be private (use `__balance`). Prevent negative withdrawals.

````python
# YOUR CODE HERE
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.__balance = balance
    def deposit(self, amount):
        self.__balance += amount
        print("Deposit is completed. The total balance is ", self.__balance)
    def withdraw(self, amount):
        if self.__balance < amount:
            print("Insufficient Balance.")
        else:
            self.__balance -= amount
            print("Withdraw is completed. The total balance is ", self.__balance)
    def get_balance(self):
        return self.__balance
acc = BankAccount("Ram", 10000)
acc.withdraw(11000)
acc.deposit(2000)
print(acc.get_balance())
acc.withdraw(11000)
````

**Q3.** Create a base class `Shape` with a method `area()`. Create child classes `Circle` and `Triangle` that inherit from `Shape` and implement their own `area()`. Demonstrate polymorphism.

````python
# YOUR CODE HERE
class Shape:
    def area(self):
        return "Shape area"
    def peri(self):
        return "Shape perimeter"
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius**2
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height
    def area(self):
        return 0.5 * self.base * self.height
shape = Shape()
print(shape.area(), shape.peri())
circle = Circle(7)
print(circle.area(), circle.peri())
triangle = Triangle(2, 5)
print(triangle.area(), triangle.peri())
````

**Q4.** Create a `Student` class with `__str__` method. Create 5 student objects, store them in a list, and sort by average marks using a lambda.

````python
# YOUR CODE HERE
class Student:
    def __init__(self, name, marks):
        self.name = name 
        self.marks = marks
    def __str__(self):
        return f"Name : {self.name}, Marks : {self.marks}"
    def avg(self):
        return sum(self.marks)/len(self.marks)
s1 = Student("Manaswini", [20, 11, 15, 20])
s2 = Student("Raga Vyshnavi", [10, 14, 11, 19])
s3 = Student("Manogna Pranavi", [19, 12, 14, 20])
s4 = Student("Rajesh", [17, 11, 16, 20])
s5 = Student("Shiva Shankar", [18, 12, 14, 20])
l = [s1, s2, s3, s4, s5]
print("Unsorted")
for i in l:
    print(i)
l.sort(key = lambda x: x.avg())
print("\n\nSorted")
for i in l:
    print(i)
````

**Q5.** Prove that Python's built-in `list` is a class by:
1. Checking `type([1,2,3])`
2. Using `dir()` to show its methods
3. Calling `isinstance([], list)`
4. Creating your own `MyList` class that has `add()` and `size()` methods

````python
# YOUR CODE HERE
print(type([1,2,3]))
print(dir([1,2,3]))
print(isinstance([], list))

class MyList:
    def __init__(self, l):
        self.l = l
    def add(self):
        s = 0
        for i in self.l:
            s += i
        return s
    def size(self):
        c = 0
        for i in self.l:
            c += 1
        return c
    def __str__(self):
        return str(self.l)
lst = MyList([1,2,3,4,5,6,7])
print(lst)
print(lst.add())
print(lst.size())
````

**Q6.** Create a `Library` class that manages a collection of books. It should support:
- `add_book(title, author)`
- `remove_book(title)`
- `search_by_author(author)` → returns list of books
- `show_all()` → prints all books

Use a list of dictionaries internally.

````python
# YOUR CODE HERE
class Library:
    def __init__(self):
        self.books = []
    def add_book(self, title, author):
        book = {"Title" : title, "Author" : author}
        self.books.append(book)
        print("Book Added")
    def remove_book(self, title):
        for i in self.books:
            if i['Title'] == title:
                self.books.remove(i)
                print("Book removed")
                return
        print("Book not found")
    def search_by_author(self, author):
        ans = []
        for i in self.books:
            if i["Author"] == author:
                ans.append(i)
        return ans
    def show_all(self):
        for i in self.books:
            print(f'"Title" : {i["Title"]}, "Author" : {i["Author"]}')
lib = Library()
lib.add_book("The Alchemist", "Paulo Coelho")
lib.add_book("Harry Potter and the Sorcerer's Stone", "J.K. Rowling")
lib.add_book("The Great Gatsby", "F. Scott Fitzgerald")
lib.remove_book("The Great Gatsby")
lib.remove_book("To Kill a Mockingbird")
lib.add_book("Eleven Minutes", "Paulo Coelho")
print(lib.search_by_author("Paulo Coelho"))
lib.show_all()
````

## 🎯 Bonus Challenge (Optional)

Create a `Calculator` class that:
- Stores a running result (starts at 0)
- Has methods: `add()`, `subtract()`, `multiply()`, `divide()`, `reset()`, `get_result()`
- Supports chaining: `calc.add(5).multiply(3).subtract(2).get_result()` → 13
- Has a `__str__` that shows the current result

Hint: return `self` from each method to enable chaining.

````python
# YOUR CODE HERE
class Calculator:
    def __init__(self):
        self.result = 0
    def add(self, val):
        self.result += val
        return self
    def subtract(self, val):
        self.result -= val
        return self
    def multiply(self, val):
        self.result *= val
        return self
    def divide(self, val):
        if val != 0:
            self.result /= val
        else:
            print("Division by 0 not possible")
        return self
    def reset(self):
        self.result = 0
        return self
    def get_result(self):
        return self.result
    def __str__(self):
        return f"Current Result : {str(self.result)}"
        
calc = Calculator()
res = calc.add(5).subtract(3).multiply(7).get_result()    
print(res)
print(calc)
res = calc.add(5).subtract(3).multiply(7).get_result()   
print(res)
res = calc.reset()
print(res)
````
